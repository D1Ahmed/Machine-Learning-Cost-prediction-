```python
import numpy as np
import matplotlib.pyplot as plt
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])      
```

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
neg=y_train==0
pos=y_train==1

ax.scatter(X_train[pos,0],X_train[pos,1],marker='x',s=100,c='red',label='y=1')
ax.scatter(X_train[neg,0],X_train[neg,1],marker='o',s=100,c='blue',label='y=0')
ax.axis([0,4,0,3])


plt.show()
```

```python
#sigmoid function:
def sigmoid(z):
    g=1/(1+np.exp(-z))
    return g
    
```

```python
#logistic Cost
def logisticCost(x,y,w,b):
    m=x.shape[0]
    cost=0.0
    for i in range(m):
        fwb=np.dot(x[i],w)+b #prediction linear regression model
        g=sigmoid(fwb) #that prediction goes to sigmoid to tell us probability of rightness/wrongness
        cost=cost+ ((y[i]*np.log(g))+((1-y[i])* np.log (1-g))) #calculate loss at each data, summing up to get total cost
    cost=-(cost/m)
    return cost
    
w_tmp = np.array([1,1])
b_tmp = -2.5   
c= logisticCost(X_train,y_train,w_tmp,b_tmp)
print(c)
```
